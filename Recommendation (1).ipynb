```python
from matplotlib.pyplot import imread, imshow
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.imagenet_utils import decode_predictions, preprocess_input
from tensorflow.keras.models import load_model
import cv2
import numpy as np
import os


new_model = load_model('classyboy')
```

```python
def harvest_recommend_system(folder):    
    dataset = os.listdir(folder)
    len_dst = len(dataset)
    count_wf = 0

    for img in dataset:
        path = f"{folder}\\{img}"
        img = cv2.imread(path)
        img = cv2.resize(img, (224, 224))
        yhat=new_model.predict(np.expand_dims(img/255,0))
        if yhat[0][1] < yhat[0][0]:
            count_wf += 1
    percent = int((count_wf/len_dst)*100)
    
    if percent < 5:
        return f"\n\nDate: {folder[-10:]} \nNo Forage in budding stage. \n\nRecommendation: WAIT"
    elif percent < 10 and percent >= 5:
        return f"\n\nDate: {folder[-10:]} \nAround {percent}% Forage in budding stage. \n\nRecommendation: WAIT"
    elif percent < 20 and percent >= 10:
        return f"\n\nDate: {folder[-10:]} \nAround {percent}% Forage in budding stage. \n\nRecommendation: WAIT"
    elif percent <= 30 and percent >= 20 and percent != 25:
        return f"\n\nDate: {folder[-10:]} \nAround {percent}% Forage in budding stage. \n\nRecommendation: HARVEST"
    elif percent == 25:
        return f"\n\nDate: {folder[-10:]} \nAround 25% Forage in budding stage. \nOptimal harvesting time for best Digestibility. \n\nRecommendation: HARVEST"
    else:
        return f"\n\nDate: {folder[-10:]} \nAround {percent}% Forage in budding stage. \nDigestibility considerably reduced. \n\nRecommendation: HARVEST"
```
